```python
def round_scores(student_scores):
    """
    :param student_scores: list of student exam scores as float or int.
    :return: list of student scores *rounded* to nearest integer value.
    """

    output_list = []
    student_scores.reverse()
    for data in student_scores:
        output_list.append(round(data))
    return output_list

print(round_scores([90.33, 40.5, 55.44, 70.05, 30.55, 25.45, 80.45, 95.3, 38.7, 40.3]))

def count_failed_students(student_scores):
    """
    :param student_scores: list of integer student scores.
    :return: integer count of student scores at or below 40.
    """

    output = 0
    for data in student_scores:
        if data <= 40:
            output = output + 1
    return output 

print(count_failed_students([90,40,55,70,30,25,80,95,38,40]))


def above_threshold(student_scores, threshold):
    """
    :param student_scores: list of integer scores
    :param threshold :  integer
    :return: list of integer scores that are at or above the "best" threshold.
    """

    output_list = []
    for data in student_scores:
        if data >= threshold:
            output_list.append(data)
    return output_list

print(above_threshold([90,40,55,70,30,68,70,75,83,96], 75))


def letter_grades(highest):
    """
    :param highest: integer of highest exam score.
    :return: list of integer score thresholds for each F-A letter grades.
    """

    threshold = 41
    return list(range(threshold, highest, round((highest - threshold) / 4)))

print(letter_grades(100))

def student_ranking(student_scores, student_names):
    """
     :param student_scores: list of scores in descending order.
     :param student_names: list of names in descending order by exam score.
     :return: list of strings in format ["<rank>. <student name>: <score>"].
     """

    output_list = []
    rank = 0
    for index, data in enumerate(student_names):
        rank = rank + 1
        input = str(rank) + ". " + student_names[index] + ": " + str(student_scores[index])
        output_list.append(input)
    return output_list

print(student_ranking([100, 99, 90, 84, 66, 53, 47], ['Joci', 'Sara','Kora','Jan','John','Bern', 'Fred']))


def perfect_score(student_info):
    """
    :param student_info: list of [<student name>, <score>] lists
    :return: first `[<student name>, 100]` or `[]` if no student score of 100 is found.
    """

    for data in student_info:
        print(data[1])
        if data[1] == 100:
            return data 
    return []

print(perfect_score([["Charles", 90], ["Tony", 80], ["Alex", 100]]))
print(perfect_score([["Charles", 90], ["Tony", 80]]))
def round_scores(student_scores):
    """
    :param student_scores: list of student exam scores as float or int.
    :return: list of student scores *rounded* to nearest integer value.
    """

    output_list = []
    student_scores.reverse()
    for data in student_scores:
        output_list.append(round(data))
    return output_list

print(round_scores([90.33, 40.5, 55.44, 70.05, 30.55, 25.45, 80.45, 95.3, 38.7, 40.3]))

def count_failed_students(student_scores):
    """
    :param student_scores: list of integer student scores.
    :return: integer count of student scores at or below 40.
    """

    output = 0
    for data in student_scores:
        if data <= 40:
            output = output + 1
    return output 

print(count_failed_students([90,40,55,70,30,25,80,95,38,40]))


def above_threshold(student_scores, threshold):
    """
    :param student_scores: list of integer scores
    :param threshold :  integer
    :return: list of integer scores that are at or above the "best" threshold.
    """

    output_list = []
    for data in student_scores:
        if data >= threshold:
            output_list.append(data)
    return output_list

print(above_threshold([90,40,55,70,30,68,70,75,83,96], 75))


def letter_grades(highest):
    """
    :param highest: integer of highest exam score.
    :return: list of integer score thresholds for each F-A letter grades.
    """

    threshold = 41
    return list(range(threshold, highest, round((highest - threshold) / 4)))

print(letter_grades(100))

def student_ranking(student_scores, student_names):
    """
     :param student_scores: list of scores in descending order.
     :param student_names: list of names in descending order by exam score.
     :return: list of strings in format ["<rank>. <student name>: <score>"].
     """

    output_list = []
    rank = 0
    for index, data in enumerate(student_names):
        rank = rank + 1
        input = str(rank) + ". " + student_names[index] + ": " + str(student_scores[index])
        output_list.append(input)
    return output_list

print(student_ranking([100, 99, 90, 84, 66, 53, 47], ['Joci', 'Sara','Kora','Jan','John','Bern', 'Fred']))


def perfect_score(student_info):
    """
    :param student_info: list of [<student name>, <score>] lists
    :return: first `[<student name>, 100]` or `[]` if no student score of 100 is found.
    """

    for data in student_info:
        print(data[1])
        if data[1] == 100:
            return data 
    return []

print(perfect_score([["Charles", 90], ["Tony", 80], ["Alex", 100]]))
print(perfect_score([["Charles", 90], ["Tony", 80]]))
def round_scores(student_scores):
    """
    :param student_scores: list of student exam scores as float or int.
    :return: list of student scores *rounded* to nearest integer value.
    """

    output_list = []
    student_scores.reverse()
    for data in student_scores:
        output_list.append(round(data))
    return output_list

print(round_scores([90.33, 40.5, 55.44, 70.05, 30.55, 25.45, 80.45, 95.3, 38.7, 40.3]))

def count_failed_students(student_scores):
    """
    :param student_scores: list of integer student scores.
    :return: integer count of student scores at or below 40.
    """

    output = 0
    for data in student_scores:
        if data <= 40:
            output = output + 1
    return output 

print(count_failed_students([90,40,55,70,30,25,80,95,38,40]))


def above_threshold(student_scores, threshold):
    """
    :param student_scores: list of integer scores
    :param threshold :  integer
    :return: list of integer scores that are at or above the "best" threshold.
    """

    output_list = []
    for data in student_scores:
        if data >= threshold:
            output_list.append(data)
    return output_list

print(above_threshold([90,40,55,70,30,68,70,75,83,96], 75))


def letter_grades(highest):
    """
    :param highest: integer of highest exam score.
    :return: list of integer score thresholds for each F-A letter grades.
    """

    threshold = 41
    return list(range(threshold, highest, round((highest - threshold) / 4)))

print(letter_grades(100))
print(letter_grades(80))

def student_ranking(student_scores, student_names):
    """
     :param student_scores: list of scores in descending order.
     :param student_names: list of names in descending order by exam score.
     :return: list of strings in format ["<rank>. <student name>: <score>"].
     """

    output_list = []
    rank = 0
    for index, data in enumerate(student_names):
        rank = rank + 1
        input = str(rank) + ". " + student_names[index] + ": " + str(student_scores[index])
        output_list.append(input)
    return output_list

print(student_ranking([100, 99, 90, 84, 66, 53, 47], ['Joci', 'Sara','Kora','Jan','John','Bern', 'Fred']))


def perfect_score(student_info):
    """
    :param student_info: list of [<student name>, <score>] lists
    :return: first `[<student name>, 100]` or `[]` if no student score of 100 is found.
    """

    for data in student_info:
        print(data[1])
        if data[1] == 100:
            return data 
    return []

print(perfect_score([["Charles", 90], ["Tony", 80], ["Alex", 100]]))
print(perfect_score([["Charles", 90], ["Tony", 80]]))
```
