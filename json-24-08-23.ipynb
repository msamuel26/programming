Perintah dibawah untuk membuka file, "r" untuk membuka dengan konsep "read"

```python
f = open("trial-migrasi-cmsdb-hris.json", "r")
```

Sekarang membaca file

```python
data = f.read()
print(data)
```

```python
print(type(data))
```

Dari hasil diatas, dapat diketahui bahwa value dari key "TableMappings" dan "ReplicationTaskSettings" bertipe string dan 
hasilnya tidak dapat mudah dibaca, 
Maka valuenya perlu untuk diubah ke JSON untuk dapat mengambil value dari key di atas

```python
import json

d_data = json.loads(data)
tm = d_data['TableMappings']
print(tm)
```

```python
print(type(tm))
```

```python
d_tm = json.loads(tm)
print(type(d_tm))
```

Dibawah ini bertujuan untuk hasil "TableMappings" yang telah di ubah ke format JSON ("d_tm") ditaruh ke "d_data",
kemudian dari hasil tersebut dictionary bernama "d_data" dilakukan perintah "dumps" dan hasilnya ditaruh ke variabel "result"

```python
d_data['TableMappings'] = d_tm
result = json.dumps(d_data, indent=2, default=str)
print(result)
```

```python
print(type(result))
```

```python
rts = d_data['ReplicationTaskSettings']
d_rts = json.loads(rts) 

d_data['ReplicationTaskSettings'] = d_rts
result1 = json.dumps(d_data, indent=2, default=str)

print(result1)
```
